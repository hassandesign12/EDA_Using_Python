Question: Which airline had the highest percentage of delayed or cancelled flights in 2015?

```python
flights_2015 = flights_data[flights_data['YEAR'] == 2015]
delayed_or_cancelled = flights_2015[(flights_2015['ARRIVAL_DELAY'] > 0) | (flights_2015['CANCELLED'] == 1)]
percentage_delayed_or_cancelled = delayed_or_cancelled.groupby('AIRLINE').size() / flights_2015.groupby('AIRLINE').size() * 100
airline_highest_delay_percentage = percentage_delayed_or_cancelled.idxmax()
airline_highest_delay_percentage
```

crearing graphs

```python
import matplotlib.pyplot as plt

# Plotting the bar chart
plt.figure(figsize=(10, 6))
percentage_delayed_or_cancelled.plot(kind='bar', color='skyblue')
plt.title('Percentage of Delayed or Cancelled Flights by Airline in 2015')
plt.xlabel('Airline')
plt.ylabel('Percentage')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Q2:Question: Which airports had the most flight cancellations?


```python
flights_data['ORIGIN_AIRPORT'].unique()
cancelled_flights = flights_data[flights_data['CANCELLED'] == 1]
cancelled_by_airport = cancelled_flights.groupby('ORIGIN_AIRPORT').size()
airports_most_cancellations = cancelled_by_airport.idxmax()
num_cancellations = cancelled_by_airport.max()
airports_most_cancellations

```

```python
top_airports = cancelled_by_airport.nlargest(10)

plt.figure(figsize=(12, 6))
top_airports.plot(kind='bar', color='skyblue')
plt.title(f'Top {10} Airports with the Most Flight Cancellations')
plt.xlabel('Airport')
plt.ylabel('Number of Cancellations')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Question 3: Are there any geographical patterns in flight delays? Do certain regions or airports experience more delays than others?

```python
import seaborn as sns
delay_statistics = flights_data.groupby('ORIGIN_AIRPORT')[['DEPARTURE_DELAY', 'ARRIVAL_DELAY']].mean()

plt.figure(figsize=(10, 20))
sns.heatmap(delay_statistics, annot=False, cmap='coolwarm', fmt=".1f")
plt.title('Average Flight Delay by Airport or Region')
plt.xlabel('Delay Type')
plt.ylabel('Airport or Region')
plt.show()
```

Q4: Which routes (origin-destination pairs) have the highest average delays?

```python
route_delays = flights_data.groupby(['ORIGIN_AIRPORT', 'DESTINATION_AIRPORT'])['ARRIVAL_DELAY'].mean().reset_index()
top_route_delays = route_delays.sort_values(by='ARRIVAL_DELAY', ascending=False).head(10)
top_route_delays
```

```python
plt.figure(figsize=(14, 8))
plt.barh(top_route_delays['ORIGIN_AIRPORT'] + " to " + top_route_delays['DESTINATION_AIRPORT'], top_route_delays['ARRIVAL_DELAY'], color='skyblue')
plt.xlabel('Average Arrival Delay (minutes)')
plt.title('Top 10 Routes with the Highest Average Delays')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest delays at the top
plt.tight_layout()
plt.show()
```
